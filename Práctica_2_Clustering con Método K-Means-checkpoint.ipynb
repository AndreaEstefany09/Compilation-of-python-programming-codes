```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

%matplotlib inline
```

## <p style="color:blue;">Importar datos

```python
vinos=pd.read_csv('C:/Users/LENOVO/Desktop/Diplomado/Modulo 3/caracteristicas de vinos.csv',
                  engine='python')
```

### <p style="color:purple;">Ver características del objeto vinos

```python
vinos.info()
```

Mostrar una vista previa 

```python
vinos.head()
```

Quitar la columna "Vino" del análisis de características

```python
vinos_variables=vinos.drop(('Vino'), axis=1)
```

```python
vinos_variables.describe()
```

### <p style="color:purple;">Normalizamos los valores de las características

```python
vinos_norm=(vinos_variables-vinos_variables.min())/(vinos_variables.max()-vinos_variables.min())
vinos_norm
```

```python
vinos_norm.describe()
```

###  <p style="color:purple;">Búsqueda de la cantidad óptima de clusters 

Calculando qué tan similares son los individuos dentro de los clusters

```python
#To measure this distance,the sum of the squares within each group is used, also called WCSS for its acronym in english (Within-Cluster Sum of Squares)
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, max_iter = 300)
    kmeans.fit(vinos_norm)
    wcss.append(kmeans.inertia_)
```

Graficando los resultados de WCSS para formar el Codo de Jambú

```python
plt.plot(range(1, 11), wcss)
plt.title("Codo de Jambú")
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
```

   ### <p style="color:purple;">Aplicando el método K-means a la base de datos

```python
clustering = KMeans(n_clusters = 3, max_iter = 300) #Crea el modelo 
clustering.fit(vinos_norm) #Aplica el modelo a la base de datos
```

### <p style="color:purple;">Agregando la clasificación al archivo original

```python
vinos['KMeans_Clusters'] = clustering.labels_
vinos.head()
```

### <p style="color:purple;">Visualizando los clusters que se formaron 

Aplicaremos el análisis de componentes principales para darnos una idea de como se formaron los clusters

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_vinos = pca.fit_transform(vinos_norm)
pca_vinos_df = pd.DataFrame (data = pca_vinos, columns = ['Componente_1', 'Componente_2'])
pca_nombres_vinos = pd.concat([pca_vinos_df, vinos[['KMeans_Clusters']]], axis=1)

pca_nombres_vinos
```

```python
fig = plt.figure(figsize = (6,6))

ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Componente 1', fontsize = 15)
ax.set_ylabel('Componente 2', fontsize = 15)
ax.set_title('Componentes Principales', fontsize = 20)

color_theme = np.array(["blue", "green", "orange"])
ax.scatter(x = pca_nombres_vinos.Componente_1, y = pca_nombres_vinos.Componente_2,
            c=color_theme[pca_nombres_vinos.KMeans_Clusters], s = 50)
plt.show()
```

    En la gráfica anterior se pueden observar que hay 3 clusters, donde se representa en cada punto a uno de los vinos y los colores al clusters al que pertenece.
    Se puede observar que los puntos si están en el cluster de acuerdo a su color y no mezclados de alguna forma incorrecta.
